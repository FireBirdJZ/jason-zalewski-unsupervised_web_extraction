```python
from bs4 import BeautifulSoup
import requests

# Make a request to the webpage
url = 'https://cs.illinois.edu/about/people/all-faculty/jeffe'
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find the element containing the "Education" heading
heading = soup.find('h2', text='Education')

# Find all the elements after the "Education" heading
elements = []
current = heading.next_sibling
while current is not None:
    elements.append(current)
    current = current.next_sibling

# Print the text content of the extracted elements
for element in elements:
    print(element.get_text(strip=True))
```

```python
from lxml import html
import requests

# Make a request to the webpage
url = 'https://cs.illinois.edu/about/people/all-faculty/jeffe'
response = requests.get(url)

# Parse the HTML content using lxml
tree = html.fromstring(response.content)

# Find the "Education" header using xpath
education_header = tree.xpath('//h2[text()="Education"]')[0]

# Get all the elements after the "Education" header
elements = education_header.xpath('following-sibling::*')

# Extract the text from the elements
for element in elements:
    print(element.text_content().strip())
```
